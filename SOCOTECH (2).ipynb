```python
import os
import pandas as pd

# Liste des identifiants de dossiers
departements = ["04", "05", "06", "13", "34", "69", "75", "93"]


df_list = []


def process_directory(directory_path):
    try:
        rel = pd.read_csv(os.path.join(directory_path, "rel_batiment_groupe_dpe_logement.csv"), encoding="utf-8")
        dpe = pd.read_csv(os.path.join(directory_path, "dpe_logement.csv"), encoding="utf-8", low_memory=False)
        batiment = pd.read_csv(os.path.join(directory_path, "batiment_groupe.csv"), encoding="utf-8")
        bat_cstr = pd.read_csv(os.path.join(directory_path, "batiment_construction.csv"), encoding="utf-8")
        adresse = pd.read_csv(os.path.join(directory_path, "adresse.csv"), encoding="utf-8")
        usage = pd.read_csv(os.path.join(directory_path, "batiment_groupe_synthese_propriete_usage.csv"), encoding="utf-8")

        # Jointures
        df = pd.merge(rel, dpe, on="identifiant_dpe", how="left")
        df = pd.merge(df, batiment[['batiment_groupe_id', 'libelle_commune_insee']], on="batiment_groupe_id", how="left")
        df = pd.merge(df, bat_cstr[['batiment_groupe_id', 'hauteur', 'altitude_sol']], on="batiment_groupe_id", how="left")
        df = pd.merge(df, adresse[['cle_interop_adr', 'numero', 'type_voie', 'nom_voie', 'code_postal', 'libelle_commune']], on="cle_interop_adr", how="left")
        df = pd.merge(df, usage[['batiment_groupe_id', 'usage_principal_bdnb_open', 'code_departement_insee']], on="batiment_groupe_id", how="left")

        # Colonnes utiles
        colonnes_utiles = [
            'identifiant_dpe', 'batiment_groupe_id',
            'classe_bilan_dpe', 'annee_construction_dpe', 'surface_habitable_logement',
            'type_batiment_dpe',
            'numero', 'type_voie', 'nom_voie', 'code_postal', 'libelle_commune',
            'usage_principal_bdnb_open', 'code_departement_insee'
        ]

        df_final = df[colonnes_utiles].copy()

        df_final.rename(columns={
            'surface_habitable_logement': 'surface_logement_m2',
            'code_departement_insee': 'departement',
            'libelle_commune': 'commune_adresse',
            'libelle_commune_insee': 'commune_insee'
        }, inplace=True)

        # Nettoyage + échantillonnage
        df_final.dropna(inplace=True)
        df_final.drop_duplicates(inplace=True)
        df_final_sampled = df_final.sample(n=10000, random_state=42)

        return df_final_sampled

    except Exception as e:
        print(f"Erreur pour le dossier {directory_path} : {e}")
        return None


for dep in departements:
    folder_path = f'/content/data{dep}'
    df_processed = process_directory(folder_path)
    if df_processed is not None:
        df_list.append(df_processed)


if df_list:
    df_global = pd.concat(df_list, ignore_index=True)
    df_global.to_csv("/content/global_bdnb_data_sampled.csv", index=False)
    print("✅ Fichier global sauvegardé dans /content/global_bdnb_data_sampled.csv")
else:
    print("❌ Aucun fichier n'a pu être traité.")
```

```python
import pandas as pd
import numpy as np
import faiss

df = pd.read_parquet('/content/bdnb_clean.parquet')
texts = df['metadata_text'].tolist()  # Nos textes métier prétraités
from sentence_transformers import SentenceTransformer

# Chargement du modèle léger mais performant
model = SentenceTransformer('all-MiniLM-L6-v2', device='cpu')

# Génération des embeddings (batch processing pour optimisation)
batch_size = 128
embeddings = model.encode(texts, batch_size=batch_size, show_progress_bar=True, convert_to_numpy=True)
embeddings = embeddings.astype('float32')
```

```python
# création de l'index FAISS, l'ajout des vecteurs dans cet index, et enfin la
# sauvegarde de cet index pour l'utiliser plus tard sans devoir recréer les embeddings.
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)

# Normalisation des vecteurs pour améliorer les performances
# Ajout des vecteurs à l'index
faiss.normalize_L2(embeddings)
index.add(embeddings)

# Sauvegarde de l'index pour l utiliser apres
faiss.write_index(index, '/content/bdnb_faiss.index')
```

```python
def rag_search(query, k=5):
    # Encodage de la requête
    query_embedding = model.encode([query], convert_to_numpy=True).astype('float32')
    faiss.normalize_L2(query_embedding)
    # recherche dans l'index FAISS pour trouver les k vecteurs les plus proches de la requête donnée.
    distances, indices = index.search(query_embedding, k)
    # Récupération des résultats
    results = df.iloc[indices[0]].copy()
    # Calcul du score de similarité
    results['similarity_score'] = 1 - distances[0]
    return results.sort_values('similarity_score', ascending=False)
```

```python
# Test avec une question métier , le model est sensible à certains variables comme le code postale , la ville , la classe energetique .
question = "Quels sont les 10 quartiers de Marseille avec le plus de passoires thermiques "
results = rag_search(question, k=3)
print(f"Résultats pour : {question}\n")
for idx, row in results.iterrows():
    print(f"→ {row['metadata_text']} (Score: {row['similarity_score']:.2f})")
```

```python
from transformers import pipeline

# Chargement du modèle T5 , un model simple mais pas tres performant
rag_generator = pipeline("text2text-generation", model="google/flan-t5-base")

def generate_answer(question, retrieved_texts):
    # On formate le contexte avec les métadonnées récupérées (limitées à 3-5 pour ne pas saturer le modèle)
    context = "\n".join(retrieved_texts[:5])
    prompt = f"Question : {question}\n\nContexte :\n{context}\n\nRéponds de manière claire et concise à la question."

    # Génération de la réponse
    output = rag_generator(prompt, max_new_tokens=150, do_sample=False)
    return output[0]['generated_text']
```

```python
# Exemple , permiere tentative
question = "Quels sont les 10 quartiers de Marseille avec le plus de passoires thermiques "
results = rag_search(question, k=3)
retrieved_texts = results['metadata_text'].tolist()

answer = generate_answer(question, retrieved_texts)
print("Réponse générée par le LLM :")
print(answer)
```
